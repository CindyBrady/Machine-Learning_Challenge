```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#  Merge txt Files

```python
df1 = pd.read_csv("Resources/2012-13/raw_scores.txt")
df2 = pd.read_csv("Resources/2013-14/raw_scores.txt")
df3 = pd.read_csv("Resources/2014-15/raw_scores.txt")
df4 = pd.read_csv("Resources/2015-16/raw_scores.txt")
df5 = pd.read_csv("Resources/2016-17/raw_scores.txt")
df6 = pd.read_csv("Resources/2017-18/raw_scores.txt")
df7 = pd.read_csv("Resources/2018-19/raw_scores.txt")
```

```python
# Combine csv files into dataframe
frames = [df1, df2, df3, df4, df5, df6, df7]
df = pd.concat(frames).drop_duplicates()
```

```python
# Export dataframe to csv
df.to_csv("Resources/raw_scores_combined.csv", index=False)
```

```python
raw_combined_df = pd.read_csv("Resources/raw_scores_combined.csv")
raw_combined_df = raw_combined_df.loc[
    (raw_combined_df["TEAM_ABBREVIATION"] != "EST")&
    (raw_combined_df["TEAM_ABBREVIATION"] != "WST")&
    (raw_combined_df["TEAM_ABBREVIATION"] != "NOH")&
    (raw_combined_df["TEAM_ABBREVIATION"] != "LBN")&
    (raw_combined_df["TEAM_ABBREVIATION"] != "STP")
                                     ]
raw_combined_df
```

```python
teamid_group = raw_combined_df.groupby('TEAM_ID')

teamid_count = teamid_group['TEAM_ID'].count()
teamid_count
```

```python
team_group = raw_combined_df.groupby('TEAM_ABBREVIATION')

count_teams = team_group['TEAM_ABBREVIATION'].count()
count_teams
```

```python
# Practice run to see size/layout 

team_chart = count_teams.plot(kind = 'bar', figsize = (12, 5))

team_chart.set_xlabel("Teams")
team_chart.set_ylabel("test1")

plt.show()
# plt.tight_layout()
```

```python
single_season_df.groupby("TEAM_ABBREVIATION")["PTS"].sum().plot(kind = 'bar', figsize = (12, 5), color = 'orange')
plt.title("2012-2013 Season Point Totals")
plt.xlabel("Teams")
plt.ylabel("Total Season Points")


plt.savefig("team_points_12_13.png")
plt.show()
```
