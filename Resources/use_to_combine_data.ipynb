```python
df1 = pd.read_csv("2018-19/raw_scores.txt")
df2 = pd.read_csv("2018-19/vegas_playoff.txt")
```

```python
sub_df = df2["TeamId"]
```

```python
#merged_df = df1.merge(sub_df, how="outer")
#merged_df
merged_df = df1.merge(sub_df, how="outer", left_on="TEAM_ID", right_on="TeamId")
merged_df
```
